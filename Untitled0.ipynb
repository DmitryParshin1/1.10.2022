```python
df_new = df[
    (df['Company'] == 'MSI')
    & (df['Gpu'] == 'Nvidia GeForce GTX 1050 Ti')
    & ~(df['Inches'] == 15.6)  # & (df['Inches'] != 15.6)
]

df_new['Price_euros'].min(), df_new['Price_euros'].max()
```
